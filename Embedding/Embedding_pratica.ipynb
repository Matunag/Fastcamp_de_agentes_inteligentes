```python
from qdrant_client.models import PointStruct
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("BAAI/bge-small-en-v1.5")
cursos = [
    (
        "Fundamentos da Agronomia Tropical",
        "Estudo dos sistemas de produção agrícola em clima tropical incluindo manejo do solo fertilidade irrigação e controle de pragas",
        "Agronomia"
    ),
    (
        "Introdução à Medicina Clínica",
        "Bases da anamnese exame físico raciocínio diagnóstico e princípios de conduta terapêutica em ambiente hospitalar",
        "Medicina"
    ),
    (
        "Cálculo Estrutural Aplicado",
        "Análise de esforços dimensionamento de vigas pilares e lajes segundo normas técnicas de engenharia civil",
        "Engenharia Civil"
    ),
    (
        "História Contemporânea Mundial",
        "Estudo dos principais eventos políticos sociais e econômicos do século XX e suas implicações geopolíticas",
        "História"
    ),
    (
        "Genética Molecular",
        "Estrutura e função do DNA mecanismos de replicação transcrição tradução e aplicações em biotecnologia",
        "Biologia"
    ),
    (
        "Princípios de Economia Macroeconômica",
        "Análise de inflação política monetária crescimento econômico e indicadores macroeconômicos globais",
        "Economia"
    ),
    (
        "Direito Constitucional Brasileiro",
        "Estudo da organização do Estado direitos fundamentais controle de constitucionalidade e separação de poderes",
        "Direito"
    ),
    (
        "Termodinâmica para Engenharia Mecânica",
        "Leis da termodinâmica ciclos térmicos máquinas térmicas e sistemas de geração de energia",
        "Engenharia Mecânica"
    ),
    (
        "Neuroanatomia Funcional",
        "Organização estrutural do sistema nervoso central vias neurais funções corticais e integração sensório motora",
        "Medicina"
    ),
    (
        "Produção Animal Sustentável",
        "Manejo nutricional genética aplicada bem-estar animal e sustentabilidade em sistemas pecuários",
        "Zootecnia"
    ),
    (
        "Teoria Política Moderna",
        "Estudo das ideias de contratualismo liberalismo marxismo e teorias contemporâneas do poder",
        "Ciência Política"
    ),
    (
        "Engenharia de Materiais Avançados",
        "Propriedades mecânicas térmicas e elétricas de metais polímeros cerâmicas e compósitos",
        "Engenharia de Materiais"
    ),
    (
        "Epidemiologia e Saúde Pública",
        "Métodos de investigação epidemiológica análise estatística de surtos e políticas de prevenção em saúde coletiva",
        "Saúde Pública"
    ),
    (
        "Arqueologia das Civilizações Antigas",
        "Estudo de métodos arqueológicos análise de artefatos e reconstrução histórica de sociedades antigas",
        "Arqueologia"
    ),
    (
        "Hidrologia e Recursos Hídricos",
        "Ciclo hidrológico modelagem de bacias hidrográficas gestão de recursos hídricos e impacto ambiental",
        "Engenharia Ambiental"
    ),
    (
        "Psicologia do Desenvolvimento",
        "Processos cognitivos emocionais e sociais ao longo da infância adolescência e vida adulta",
        "Psicologia"
    ),
    (
        "Farmacologia Clínica",
        "Mecanismos de ação farmacocinética interações medicamentosas e protocolos terapêuticos",
        "Farmácia"
    ),
    (
        "Topografia e Georreferenciamento",
        "Técnicas de levantamento planimétrico altimétrico uso de GPS e sistemas de informação geográfica",
        "Engenharia Cartográfica"
    ),
    (
        "História do Brasil Colonial",
        "Formação econômica política e social do Brasil entre os séculos XVI e XVIII",
        "História"
    ),
    (
        "Inteligência Artificial Aplicada",
        "Modelos de machine learning redes neurais processamento de linguagem natural e aplicações industriais",
        "Computação"
    )
]

pontos = []
embeddings = model.encode([f"{curso[0]} {curso[1]}" for curso in cursos])
for i, embedding in enumerate(embeddings):
    ponto = PointStruct(
        id=i,
        vector=embedding.tolist(),
        payload={
            "nome_curso": cursos[i][0],
            "descricao": cursos[i][1],
            "area": cursos[i][2],
        }
    )
    pontos.append(ponto)

cliente.upsert(
  collection_name="cursos",
  points=pontos,
)
```

Aqui iremos testar se nossos dados foram devidamente carregados com uma query ("cursos relacionados com matemática"). Sendo possível ver a maior parte dos cursos relacionados com matemática no topo.

```python
query = "cursos relacionados com matemática"
queries = model.encode(query).tolist()

resposta = cliente.query_points(
    collection_name="cursos",
    query=queries,
    with_payload=True,
    limit=5
)

for result in resposta.points:
    print(f"Curso: {result.payload.get('nome_curso', 'N/A')}")
    print(f"Score: {result.score}")
    print(f"Descrição: {result.payload['descricao'][:150]}...")
    print(f"Área: {result.payload.get('area', 'N/A')}")
    print("---")
```
